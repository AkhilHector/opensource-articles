### Implement text analysis functions to validate the summaries later

```python
class text_analyze:
    # init function
    def __init__(self, sentence_a, sentence_b):
        self.s_a = sentence_a
        self.s_b = sentence_b
    
    # tokenizer
    def tokenize(self, text):
        return re.split(r'[^0-9A-Za-z\-"_]+', text)
    
    # remove the stopwords
    def remove_stopwords(self, text):
        stmr = stemmer()
        stop_words = stopwords.words('english')
        return [w for w in [stmr.stem(words) for words in self.tokenize(text)] if w.lower() not in stop_words]
    
    # cosine similairty
    def similarity(self):
        vectorizer = TfidfVectorizer()
        sentences = tuple([' '.join(self.remove_stopwords(self.s_a))]) + tuple([' '.join(self.remove_stopwords(self.s_b))])
        matrix = vectorizer.fit_transform(sentences)
        return cosine_similarity(matrix[0:1], matrix)[0][1]
```
